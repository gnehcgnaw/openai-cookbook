###  怎样获取Java方法

####  来自你的消息:
使用python，获取Java代码中，属于方法的代码，以及方法的名称

####  来自 ChatGPT 的消息:
可以通过正则表达式匹配获取Java代码中的方法名称和方法体。

示例代码：

```python
import re

# 假设Java代码存储在文件中
with open('example.java', 'r') as f:
    java_code = f.read()

# 定义正则表达式
method_pattern = r'(?:public|private|protected)?\s*(?:void|int|float|double|char|byte|short|long|boolean)\s+(\w+)\s*\((.*?)\)(?:(?:throws\s+\w+\s*,?)*\s*)\{([\s\S]*?)\}'

# 匹配方法名称和方法体
matches = re.findall(method_pattern, java_code)

# 输出结果
for match in matches:
    method_name = match[0]
    method_body = match[2]
    print('Method name:', method_name)
    print('Method body:', method_body)
```

该代码会将Java文件中的所有方法名称和方法体输出。其中，`method_name`即为方法名称，`method_body`为方法体（包括大括号内的所有代码）。正则表达式中的`method_pattern`用于匹配方法名称和方法体，可以根据需要进行修改。

#### 来自你的消息:
method_body 要包含整个方法内容，不仅仅是大括号内的所有代码

#### 来自 ChatGPT 的消息:
我理解您的意思是希望`method_body`包括整个方法内容，包括方法名、参数和方法体。在这种情况下，可以将方法名和参数也包含在正则表达式中，然后将整个方法的内容存储在`method_body`中。

示例代码：

```python
import re

# 假设Java代码存储在文件中
with open('example.java', 'r') as f:
    java_code = f.read()

# 定义正则表达式
method_pattern = r'(?:public|private|protected)?\s*(?:void|int|float|double|char|byte|short|long|boolean)\s+(\w+)\s*\((.*?)\)(?:(?:throws\s+\w+\s*,?)*\s*)\{([\s\S]*?)\}'

# 匹配方法名称、参数和方法体
matches = re.findall(method_pattern, java_code)

# 输出结果
for match in matches:
    method_name = match[0]
    method_params = match[1]
    method_body = match[2]
    method_content = f'{method_name}({method_params})' + '{' + method_body + '}'
    print('Method content:', method_content)
```

在这个版本的代码中，`method_body`仍然包含大括号内的所有代码，但是整个方法的内容现在可以通过如下方式计算得到：

```python
method_content = f'{method_name}({method_params})' + '{' + method_body + '}'
```

其中，`f'{method_name}({method_params})'`用于构造方法名和参数部分，`'{' + method_body + '}'`则包含整个方法体，最后将两者拼接在一起即可得到整个方法的内容。

````python
import javalang
import os
# glob is used to get all files in a directory
# glob 是用来获取目录中的所有文件
# pandas is used to store the results
# pandas 用来存储结果
from glob import glob
# https://pandas.pydata.org/docs/index.html
import pandas as pd

import re



def get_functions(filepath):
    # 假设Java代码存储在文件中
    with open(filepath, 'r') as f:
        java_code = f.read()

    # 定义正则表达式
    method_pattern = r'(?:public|private|protected)?\s*(?:void|int|float|double|char|byte|short|long|boolean)\s+(\w+)\s*\((.*?)\)(?:(?:throws\s+\w+\s*,?)*\s*)\{([\s\S]*?)\}'

    # 匹配方法名称、参数和方法体
    matches = re.findall(method_pattern, java_code)

    # 输出结果
    for match in matches:
        method_name = match[0]
        method_params = match[1]
        method_body = match[2]
        method_content = f'{method_name}({method_params})' + '{' + method_body + '}'
        print('Method content:\n', method_content)
        yield {"code": method_content, "function_name": method_name, "filepath": filepath}

# get user root directory
# 获取用户根目录
root_dir = os.path.expanduser("~")
# note: for this code to work, the openai-python repo must be downloaded and placed in your root directory
# 注意：要使此代码工作，必须下载openai-python repo并将其放在您的根目录中
# path to code repository directory
code_root = root_dir + "/WorkSpaces/IntelliJIdeaPorjects/mybatis-plus-samples"
# os.walk() 方法用于通过在目录树中游走输出在目录中的文件名，向上或者向下。
# os.path.join() 方法用于将序列中的元素以指定的字符连接生成一个新的字符串。
code_files = [y for x in os.walk(code_root) for y in glob(os.path.join(x[0], '*.java'))]
print("Total number of py files:", len(code_files))

with open("data/code_files.text", 'w') as f:
    for code_file in code_files:
        f.write(code_file)
        f.write("\n")

if len(code_files) == 0:
    print("Double check that you have downloaded the openai-python repo and set the code_root variable correctly.")

all_funcs = []
# get all functions in all files
for code_file in code_files:
    funcs = list(get_functions(code_file))
    for func in funcs:
        all_funcs.append(func)

print("Total number of functions extracted:", len(all_funcs))


````

````python
# get embeddings for all functions
# 获取所有函数的嵌入
from openai.embeddings_utils import get_embedding

# pandas.DataFrame() 函数用于将字典或 Series 对象转换为 DataFrame 对象。
df = pd.DataFrame(all_funcs)
# pandas.DataFrame.apply() 函数用于对 DataFrame 的每一行或每一列应用函数。
df['code_embedding'] = df['code'].apply(lambda x: get_embedding(x, engine='text-embedding-ada-002'))
# pandas.DataFrame.to_csv() 函数用于将 DataFrame 写入 CSV 文件。
df['filepath'] = df['filepath'].apply(lambda x: x.replace(code_root, ""))
# 保存到csv文件
df.to_csv("data/code_search_mybatis-plus-samples.csv", index=False)
# pandas.DataFrame.head() 函数用于返回前 n 行。
df.head()
````

````python
# cosine_similarity 是余弦相似度 用于计算两个向量的相似度
from openai.embeddings_utils import cosine_similarity
"""
Search for functions in the code repository that are similar to a given query.
查询与给定查询类似的代码库中的函数。
"""
def search_functions(df, code_query, n=3, pprint=True, n_lines=7):
    embedding = get_embedding(code_query, engine='text-embedding-ada-002')
    # pandas.DataFrame.apply() 函数用于对 DataFrame 的每一行或每一列应用函数。
    df['similarities'] = df.code_embedding.apply(lambda x: cosine_similarity(x, embedding))
    # pandas.DataFrame.sort_values() 函数用于对 DataFrame 的列进行排序。
    res = df.sort_values('similarities', ascending=False).head(n)
    # pandas.DataFrame.iterrows() 函数用于迭代 DataFrame 的行。
    if pprint:
        for r in res.iterrows():
            print(r[1].filepath+":"+r[1].function_name + "  score=" + str(round(r[1].similarities, 3)))
            print("\n".join(r[1].code.split("\n")[:n_lines]))
            print('-'*70)
    return res

res = search_functions(df, 'testCustomSqlSegment', n=3)
````
